```python
import pandas as pd
from transformers import pipeline
import numpy as np
```

```python
sentiment_analyzer = pipeline('sentiment-analysis',
                              model = 'distilbert/distilbert-base-uncased-finetuned-sst-2-english',
                              device =0,
                              truncation = True)
```

```python
df = pd.read_excel('survey_results_translated.xlsx')
```

```python
pd.set_option('max_colwidth', True)
```

```python
df['6_translated'].head()
```

```python
text1=df['6_translated'][0]
text1
```

```python
sentiment_analyzer(text1)
```

```python
texts = df["W_in_English"].fillna("").tolist()
```

```python
results = sentiment_analyzer(texts, batch_size=16)
```

```python
df["Label_W"] = [r["label"] for r in results]
df["Score_W"] = [r["score"] for r in results]
```

```python
df.head()
```

```python
data = df.copy()
```

```python
data = data.drop(columns = ['Subject area'])
```

```python
data.head()
```

```python
data['Subject area'] = df['Subject area']
```

```python
data.to_excel("Survey_results_with_sentiment.xlsx", index=False)
```
