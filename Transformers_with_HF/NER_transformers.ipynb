# NER - named entity recognition

```python
import pandas as pd
from transformers import pipeline
```

```python
from transformers import logging
```

```python
logging.set_verbosity_error()
```

```python
ner_analyzer = pipeline(
    'ner',
    model = 'dslim/bert-base-NER',
    device=0,
    aggregation_strategy ='SIMPLE'
)
```

```python
text = "I ordered two nice jumpers yesterday from Trendyol Turkiye"
```

```python
ner_analyzer(text)
```

```python
ner_analyzer2=pipeline(
    'ner',
    model = 'dbmdz/bert-large-cased-finetuned-conll03-english',
    device=0,
    aggregation_strategy ='SIMPLE'
)
```

```python
ner_analyzer2(text)
```

```python
text2 ='I ordered two jumpers from TrendyolMilla which is famous brand in Turkiye'
```

```python
ner_analyzer(text2)
```

```python
ner_analyzer2(text2)
```
