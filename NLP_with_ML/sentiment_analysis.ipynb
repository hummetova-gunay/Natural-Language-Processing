# Sentiment Analysis 
It is used to determine the positivity or negativity of text

SA is applied to raw text - the text is not cleaned because punctuation matters, and not vectorized because word order matters

VADER - Valence Aware Dictionary and sEntiment Reasoner

```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
data=[
"When life gives you lemons, make Lemonade!",
"She bought 2 lemons for $1 at Maven Market.",
"A dozen lemons will make a gallon of lemonade. [AllRecipes]",
"lemon, Lemon, Lemons, Lemon, lemon, lemons",
"He's running to the market to get a lemon - there's a great sale today.",
"Does Maven Market carry Eureka lemons or Meyer lemons?",
"An Arnold Palmer is half lemonade, half iced tea. [Wikipedia]",
"iced tea is my favorite"
]
data_df=pd.DataFrame(data, columns=['sentences'])
data_df.head()
```

```python
df =  data_df.copy()
df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
analyzer.polarity_scores(df.sentences[0])
```

```python
df.sentences[0] = 'When life gives you lemons, make Lemonade, my sweet baby'
```

```python
df.sentences[0]
```

```python
analyzer.polarity_scores(df.sentences[0])
```

```python
# write function
def get_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    return  analyzer.polarity_scores(text)['compound']
```

```python
df.sentences.apply(get_sentiment)
```

```python
# add a new row (sentence)
df.loc[len(df)] = 'I do not like lemons at all'
```

```python
df.loc[len(df)] = 'He has been treating me like a shit'
```

```python
df.loc[len(df)] = 'You are giving me lemons as if i like them'
```

```python
# add a new column to dataframe to see sentences and sentiments side by side 
df['sentiment'] = df.sentences.apply(get_sentiment)
df
```
