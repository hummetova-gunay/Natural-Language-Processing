# Objectives
- Read in the `movie_reviews.csv` files
- Apply sentiment analysis to the movie_info column
- sort the sentiment scores to return the top 10 and bottom 10 sentiment scores and their corresponding movie titles

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
review_data = pd.read_csv('airlines_reviews.csv')
review_data.head()
```

```python
review_data.columns.tolist()
```

```python
cols_to_keep =['Airline', 'Reviews']
```

```python
df = review_data[cols_to_keep]
df.head()
```

```python
def get_sentiment(reviews):
    analyzer = SentimentIntensityAnalyzer()
    return analyzer.polarity_scores(reviews)['compound']
```

```python
df['sentiment_score'] = df.Reviews.apply(get_sentiment)
df
```

```python
top_reviews = df['sentiment_score'].sort_values().tail(10)
top_reviews
```

```python
df.loc[top_reviews.index, ['Airline', 'Reviews', 'sentiment_score']]
```

```python
bad_reviews = df['sentiment_score'].sort_values().head(10)
bad_reviews
```

```python
df.loc[bad_reviews.index, ['Airline', 'Reviews', 'sentiment_score']]
```

```python
df.columns.tolist()
```

```python
# another way of seeing top and bottom reivews
df[['Airline', 'Reviews', 'sentiment_score']].sort_values(by = 'sentiment_score', ascending=False)
```
