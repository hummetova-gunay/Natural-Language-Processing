# Data Preparation

```python
def split_comments(text):
    # Split by ' / ' and strip whitespace from each comment
    comments = [comment.strip() for comment in text.split(' / ')]
    # Remove empty strings if any
    comments = [comment for comment in comments if comment]
    
    return comments
```

# Sentiment Analysis

```python

# write function
def get_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    return  analyzer.polarity_scores(text)['compound']
```

```python
# Function to calculate mean sentiment for a list of comments
def calculate_mean_sentiment(comments_text):
    """
    Split comments, get sentiment for each, and return the mean sentiment score.
    """
    # Split the comments
    comments_list = split_comments(comments_text)
    
    # Get sentiment for each comment
    sentiments = [get_sentiment(comment) for comment in comments_list]
    
    # Calculate mean sentiment
    mean_sentiment = sum(sentiments) / len(sentiments) if sentiments else 0
    
    return mean_sentiment

# Apply the function to create a new column
df['Mean_Sentiment'] = df['Comments'].apply(calculate_mean_sentiment)
```

```python
# Add sentiment category based on mean
def categorize_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.01:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment_Category'] = df['Mean_Sentiment'].apply(categorize_sentiment)
```
