# Term Frequency - Inverse Term Frequency
$ TF- IDF =\frac{Term Count in Doc}{Total terms in doc} \times log  \frac{total documents + 1}{Docs with the term +1}$


Similar to count vectorizer but this time the document term matrix each entry represents the importance score of the terms in a document

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
data = pd.read_pickle('text_clean.pkl')
data
```

```python
tv = TfidfVectorizer(stop_words='english', ngram_range=(1,2), min_df=2)
tfidf = tv.fit_transform(data)
tfidf_df = pd.DataFrame(tfidf.toarray(), columns = tv.get_feature_names_out())
tfidf_df
```
