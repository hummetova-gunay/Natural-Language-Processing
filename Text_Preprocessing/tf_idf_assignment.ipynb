# Key Objectives
1. Vectorize the cleaned and normalized text using TF-IDF Vectorizer with the default params
2. Modify the TF-IDF Vectorizer params to reduce the number of columns
- remove stop words 
- set min_df of 10%
- set max_df of 50%
3. Using the updated TF-IDF Vectorizer, create a horizontal bar chart of the top 10 most highly weighted term (the most important terms)

```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_pickle('movie_descriptions')
df.head()
```

```python
series = df['clean_descriptions']
series
```

```python
tv = TfidfVectorizer(stop_words='english', ngram_range=(1,2), min_df = 0.1, max_df = .5)
tfidf = tv.fit_transform(series)
tfidf_df = pd.DataFrame(tfidf.toarray(), columns = tv.get_feature_names_out())
tfidf_df
```

```python
term_freq=tfidf_df.sum()
term_freq
```

```python
top_terms =  term_freq.sort_values().tail(10)
top_terms
```

```python
top_terms.plot(kind = 'barh')
```
